```python
lhs= []
rhs = []
supports = []
confidences = []
lifts = []

for result in results:
  for statistic in result.ordered_statistics:
    lhs.append(statistic.items_base)
    rhs.append(statistic.items_add)
    supports.append(result.support)
    confidences.append(statistic.confidence)
    lifts.append(statistic.lift)

res = pd.DataFrame({
    "lhs":lhs,
    "rhs":rhs,
    "support":supports,
    "confidence":confidences,
    "lift":lifts
})
res
```
